## 1. Introduction

Les réseaux sociaux constituent aujourd’hui une source importante de données permettant d’analyser les interactions entre utilisateurs. L’objectif de ce projet est d’étudier un réseau réel issu de StackOverflow afin de comprendre sa structure, ses propriétés et ses communautés.

L’analyse repose sur des méthodes issues de la théorie des graphes et du data mining, en utilisant la bibliothèque NetworkX sous Python.

## 2. Collecte et construction du réseau

Le dataset utilisé provient de StackOverflow et contient des interactions entre utilisateurs. Chaque interaction est représentée par une arête entre deux nœuds.

- **Nœuds** : utilisateurs  
- **Liens** : interactions entre utilisateurs  
- **Attributs** : timestamp des interactions  

Le réseau est construit à partir d’un fichier de données contenant les interactions utilisateur-utilisateur.

## 4. Détection des communautés

### 4.1 Méthodes utilisées

Trois algorithmes de détection de communautés ont été utilisés afin d’analyser la structure modulaire du réseau.

#### 🔹 Algorithme de Louvain

L’algorithme de Louvain est une méthode de détection de communautés basée sur l’optimisation de la modularité. La modularité mesure la qualité d’une partition en comparant la densité des liens à l’intérieur des communautés avec celle attendue dans un réseau aléatoire.

Cet algorithme fonctionne en deux étapes répétées :
- chaque nœud est déplacé vers la communauté de ses voisins si cela améliore la modularité,
- les communautés sont ensuite agrégées pour former un nouveau réseau réduit.

Louvain est connu pour sa rapidité et sa capacité à détecter des communautés de grande taille de manière efficace.

---

#### 🔹 Algorithme de Propagation des labels (Label Propagation)

L’algorithme de propagation des labels est une méthode simple et rapide basée sur la diffusion d’étiquettes dans le réseau.

Au départ, chaque nœud possède un label unique. À chaque itération :
- chaque nœud adopte le label le plus fréquent parmi ses voisins.

Ce processus continue jusqu’à stabilisation.

Cet algorithme est très rapide et ne nécessite pas de paramètre, mais ses résultats peuvent varier d’une exécution à une autre et produire des communautés moins stables.

---

#### 🔹 Algorithme K-Clique

L’algorithme K-Clique repose sur la notion de clique, c’est-à-dire un sous-graphe dans lequel tous les nœuds sont connectés entre eux.

Une communauté est définie comme un ensemble de cliques de taille k qui partagent des nœuds entre elles. Ainsi, deux cliques appartiennent à la même communauté si elles ont au moins k-1 nœuds en commun.

Cet algorithme permet de détecter des communautés très denses et fortement connectées, mais il est plus strict et peut produire un grand nombre de petites communautés.

### 4.2 Résultats

Sur un sous-graphe de 5000 nœuds :

- Louvain : 17 communautés  
- Label Propagation : 44 communautés  
- K-Clique : 147 communautés

### 4.4 Analyse comparative

Les résultats varient selon les algorithmes :

- Louvain produit des communautés bien structurées  
- Label Propagation détecte des groupes plus fins  
- K-Clique identifie des groupes très denses mais fragmentés  

Ces différences illustrent que la notion de communauté dépend fortement de la méthode utilisée.

## 5. Conclusion

L’analyse du réseau StackOverflow montre une structure typique des réseaux sociaux réels : faible densité, forte hétérogénéité des connexions et présence de communautés.

Les différentes méthodes de détection de communautés confirment l’existence d’une structure modulaire, avec des résultats variant selon les algorithmes utilisés.

Ce travail permet de mieux comprendre l’organisation des interactions dans un réseau social complexe.
